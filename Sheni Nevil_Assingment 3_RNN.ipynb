```python
# Importing the required libraries and files for the code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
from keras.layers import SimpleRNN, Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Loading the train and test dataset
data_train = pd.read_csv("Google_Stock_Price_Train.csv")
data_test = pd.read_csv("Google_Stock_Price_Test.csv")
# Displaying first 3 rows of train and test data
print(f"The first 3 rows of Train data :\n {data_train.head(3)}\n")
print(f"The first 3 rows of Test data :\n {data_test.head(3)}\n\n")

# Splitting train data as train and validation sets (70% train & 30% validation)
length_data = len(data_train)
split_ratio = 0.7
length_train = round(length_data*split_ratio)  
length_validation = length_data-length_train
print("Length of total train data :", length_data)
print("Splitted train data length :", length_train)
print("Splitted validation data length :", length_validation)

train_data = data_train[:length_train]
# Converting to date-time object
train_data['Date'] = pd.to_datetime(train_data['Date'])
train_data.head(3)

validation_data = data_train[length_train:]
# Converting to date-time object
validation_data['Date'] = pd.to_datetime(validation_data['Date'])
validation_data.head(3)

# Creating the test dataset
test_data = data_test
# Converting to date-time object
test_data['Date'] = pd.to_datetime(test_data['Date'])  
test_data.head(3)

# Creating train dataset from the splitted train 
dataset_train = train_data.Close.values
# Changing 1D to 2D array (changing shape from (1692,) to (1692,1))
dataset_train = np.reshape(dataset_train,(-1,1))
dataset_train.shape

# Creating test dataset from the splitted train
dataset_test = test_data.Close.values
# Changing 1D to 2D array
dataset_test = np.reshape(dataset_test,(-1,1))
dataset_test.shape

# Creating validation dataset from train split
dataset_val = validation_data.Close.values
# Changing 1D to 2D array (changing shape from (1692,) to (1692,1))
dataset_val = np.reshape(dataset_val,(-1,1))
dataset_val.shape

# Replacing column type
def preprocess_column(column):
    if column.dtype == 'object':
        return column.astype(str).str.replace(',', '').astype(float)
    return column
# Ensuring preprocessing is applied to the original pandas DataFrame
data_train['Close'] = preprocess_column(data_train['Close'])
data_test['Close'] = preprocess_column(data_test['Close'])
validation_data['Close'] = preprocess_column(validation_data['Close'])
# Extracting numpy arrays
dataset_train = data_train['Close'].values.reshape(-1,1)
dataset_test = data_test['Close'].values.reshape(-1,1)
dataset_val = validation_data['Close'].values.reshape(-1,1)

# Feature scaling
scaler = MinMaxScaler(feature_range = (0,1))
# scaling dataset
dataset_train_scaled = scaler.fit_transform(dataset_train)
dataset_val_scaled = scaler.fit_transform(dataset_val)
dataset_test_scaled = scaler.fit_transform(dataset_test)
plt.subplots(figsize = (5,3))
plt.plot(dataset_train_scaled)
plt.xlabel("Days")
plt.ylabel("Close Price")
plt.show()

# Creating X_train and y_train from train dataset
X_train = []
y_train = []
time_step = 2
for i in range(time_step, length_train):
    X_train.append(dataset_train_scaled[i-time_step:i,0])
    y_train.append(dataset_train_scaled[i,0])
# Converting list to array
X_train, y_train = np.array(X_train), np.array(y_train)
print("X_train before reshape :",X_train.shape)
print("y_train before reshape :",y_train.shape)
# Reshaping the datasets
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
y_train = np.reshape(y_train, (y_train.shape[0],1))
print("X_train after reshape :",X_train.shape)
print("y_train after reshape :",y_train.shape)

# Creating X_val and y_val from validation data
X_val = []
y_val = []
time_step = 2
for i in range(time_step, length_validation):
    X_val.append(dataset_val_scaled[i-time_step:i,0])
    y_val.append(dataset_val_scaled[i,0])
# Convert list to array
X_val, y_val = np.array(X_val), np.array(y_val)
print("X_val before reshape :",X_val.shape)
print("y_val before reshape :",y_val.shape)
# Reshaping the datasets
X_val = np.reshape(X_val, (X_val.shape[0], X_val.shape[1],1))
y_val = np.reshape(y_val, (y_val.shape[0],1))
print("X_val after reshape :",X_val.shape)
print("y_val after reshape :",y_val.shape)

# Creating X_test and y_test from the test data
X_test = []
y_test = []
time_step = 2
length_test = len(data_test)
for i in range(time_step, length_test):
    X_test.append(dataset_test_scaled[i-time_step:i,0])
    y_test.append(dataset_test_scaled[i,0])
# Convert list to array
X_test, y_test = np.array(X_test), np.array(y_test)
print("X_test before reshape :",X_test.shape)
print("y_test before reshape :",y_test.shape)
# Reshaping the datasets
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))
y_test = np.reshape(y_test, (y_test.shape[0],1))
print("X_test after reshape :",X_test.shape)
print("y_test after reshape :",y_test.shape)

# Defining function to create and train RNN model
def train_rnn_model(X_train,y_train,rnn_units=[128,64,32],dropout_rate=0.2,epochs=100,batch_size=10):
    model_rnn = Sequential()
    # Adding 1st layer with return_sequences=True
    model_rnn.add(SimpleRNN(rnn_units[0], return_sequences=True, input_shape=(X_train.shape[1], 1)))
    # Adding remaining layers
    for units in rnn_units[1:]:
        model_rnn.add(SimpleRNN(units, return_sequences=True))
    # Adding last layer with return_sequences=False
    model_rnn.add(SimpleRNN(rnn_units[-1], return_sequences=False))
    # Adding Dropout layer to prevent overfitting
    model_rnn.add(Dropout(dropout_rate))
    # Adding Dense output layer with 1 neuron for regression output
    model_rnn.add(Dense(1))
    # Compiling using Adam optimizer and MSE loss
    model_rnn.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    # Training model on training dataset
    history_rnn = model_rnn.fit(X_train, y_train, epochs=epochs, batch_size=batch_size)
    return model_rnn, history_rnn

def evaluate_model_performance(y_true, y_pred):
    # Calculating Mean Squared Error (MSE)
    mse = mean_squared_error(y_true, y_pred)
    print(f"Mean Squared Error(MSE) : {mse}")
    # Calculating Mean Absolute Error (MAE)
    mae = mean_absolute_error(y_true, y_pred)
    print(f"Mean Absolute Error (MAE) : {mae}")
    # Calculating Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    print(f"Root Mean Squared Error (RMSE) : {rmse}")
    # Defining tolerance level
    tolerance = 0.01
    # Calculating percentage difference between actual and predicted values
    percentage_difference = np.abs((y_pred-y_true)/y_true)
    # Checking if predictions are within tolerance
    within_tolerance = percentage_difference < tolerance
    # Calculating accuracy as a percentage of predictions within the tolerance
    accuracy = np.mean(within_tolerance) * 100
    print(f"Accuracy (within {tolerance * 100}% tolerance) : {accuracy:.2f}%")

def evaluate_predictions(model, X_data, y_data, scaler, data_type="Train"):
    # Predicting the data
    predictions = model.predict(X_data)
    # Inverse transforming predicted and actual values
    predictions = scaler.inverse_transform(predictions)
     # Ensuring actual values are also inverse transformed
    y_data = scaler.inverse_transform(y_data)   
    # Plotting predictions vs actual values
    plt.subplots(figsize=(5,3))
    plt.plot(predictions, label=f"y_pred_of_{data_type.lower()}", color="orange")
    plt.plot(y_data, label=f"y_{data_type.lower()}", color="g")
    plt.xlabel("Days")
    plt.ylabel("Open price")
    plt.title(f"RNN Model: Predicting -> input X_{data_type.lower()} vs y_{data_type.lower()}")
    plt.legend()
    plt.show()
    # Evaluating model performance
    evaluate_model_performance(y_data, predictions)

model_rnn,history_rnn = train_rnn_model(X_train,y_train,rnn_units=[128,64,32],dropout_rate=0.2,epochs=50,batch_size=32)
plt.figure(figsize = (5,3))
plt.plot(history_rnn.history["loss"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("RNN Model : Accuracy vs Epoch")
plt.show()

evaluate_predictions(model_rnn, X_train, y_train, scaler, data_type="Train")
evaluate_predictions(model_rnn, X_val, y_val, scaler, data_type="Validation")
evaluate_predictions(model_rnn, X_test, y_test, scaler, data_type="Test")

# With 2 layers 64 and 32 units
model_rnn,history_rnn = train_rnn_model(X_train,y_train,rnn_units=[64,32],dropout_rate=0.2,epochs=50,batch_size=32)
evaluate_predictions(model_rnn, X_train, y_train, scaler, data_type="Test")
evaluate_predictions(model_rnn, X_val, y_val, scaler, data_type="Test")
evaluate_predictions(model_rnn, X_test, y_test, scaler, data_type="Test")

# With 2 layers 64 and 32 units and change in batch size to 12
model_rnn,history_rnn = train_rnn_model(X_train,y_train,rnn_units=[64,32],dropout_rate=0.2,epochs=50,batch_size=12)
evaluate_predictions(model_rnn, X_train, y_train, scaler, data_type="Test")
evaluate_predictions(model_rnn, X_val, y_val, scaler, data_type="Test")
evaluate_predictions(model_rnn, X_test, y_test, scaler, data_type="Test")
```
